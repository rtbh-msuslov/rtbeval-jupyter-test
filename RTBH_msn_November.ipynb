```python
import pandas as pd
```

```python
rtbeval_countrystats = pd.read_csv("rtbeval_country_report_november.csv")
```

```python
rtbeval_cs = rtbeval_countrystats
```

```python
rtbeval_cs.loc[rtbeval_cs["domain"].str.contains("msn.com", na=False), :]
```

```python
msn_com = rtbeval_cs.loc[rtbeval_cs["domain"].str.contains("msn.com", na=False), :]
```

```python
msn_com.sort_values ('rtbhTotalWinrate', ascending = False)
```

```python
import matplotlib.pyplot as plt

# Sort data if necessary
# msn_com = msn_com.sort_values('rtbhTotalWinrate', ascending=False)

# Extract the data
countries = msn_com['country']
winrates = msn_com['rtbhTotalWinrate']

# Adjust figure size and bar height
plt.figure(figsize=(10, 12))  # Adjust figure size
bar_width = 0.8  # Adjust bar height

# Create horizontal bar chart
plt.barh(countries, winrates, height=bar_width)

# Add labels and title
plt.xlabel('RTBH Total Winrate')
plt.ylabel('Country')
plt.title('Winrate by Country')

# Set finer granularity for x-axis ticks
plt.xticks(ticks=range(0, int(max(winrates)) + 1, 5))  # Steps of 10 (adjust as needed)

# Show the plot
plt.tight_layout()
plt.show()
```
